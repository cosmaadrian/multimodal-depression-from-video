```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df_train = pd.read_csv('../data/databases/D-vlog/splits/training.csv')
df_val = pd.read_csv('../data/databases/D-vlog/splits/validation.csv')
df_test = pd.read_csv('../data/databases/D-vlog/splits/test.csv')
df_dvlog = pd.concat([df_train, df_val, df_test]).reset_index(drop = True)
df_dvlog['dataset_name'] = 'D-Vlog'

df_dvlog['label'] = df_dvlog['label'].apply(lambda x: 'Control' if x == 'normal' else 'Depression')
```

```python
df_dvlog
```

```python
df_train = pd.read_csv('../data/databases/DAIC-WOZ/splits/training.csv')
df_val = pd.read_csv('../data/databases/DAIC-WOZ/splits/validation.csv')
df_test = pd.read_csv('../data/databases/DAIC-WOZ/splits/test.csv')
df_test['PHQ8_Binary'] = df_test['PHQ_Binary']

df_woz = pd.concat([df_train, df_val, df_test]).reset_index(drop = True)
df_woz['label'] = df_woz['PHQ8_Binary'].apply(lambda x: 'Depression' if x == 1 else 'Control')
df_woz['dataset_name'] = 'DAIC-WOZ'
```

```python
df_woz
```

```python
df_train = pd.read_csv('../data/databases/E-DAIC-WOZ/splits/training.csv')
df_val = pd.read_csv('../data/databases/E-DAIC-WOZ/splits/validation.csv')
df_test = pd.read_csv('../data/databases/E-DAIC-WOZ/splits/test.csv')

df_ewoz = pd.concat([df_train, df_val, df_test]).reset_index(drop = True)
df_ewoz['dataset_name'] = 'E-DAIC-WOZ'

df_ewoz['label'] = df_ewoz['PHQ_Binary'].apply(lambda x: 'Depression' if x == 1 else 'Control')
```

```python
df_ewoz
```

```python
df_plot = pd.concat([df_dvlog[['dataset_name', 'label', 'duration']], df_woz[['dataset_name', 'label', 'duration']], df_ewoz[['dataset_name', 'label', 'duration']]]).reset_index(drop = True)
df_plot['duration'] = df_plot['duration'] / 60

df_plot
```

```python
colors = ['#E55807', "#068DA9"]
customPalette = sns.set_palette(sns.color_palette(colors))

sns.boxplot(x = 'dataset_name', y = 'duration', hue = 'label', data = df_plot)

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.tick_params(direction='out', length=10, width=1, color = '#474747', labelcolor = '#474747')
ax.tick_params(which = 'minor', direction='out', length=5, width=1, color = '#474747', labelcolor = '#474747')
ax.set_xlabel("")

ax.set_ylabel("Duration\n(minutes)", fontsize=12)

plt.title('Video Duration', fontsize = 16)
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(16)

plt.gcf().set_size_inches(6, 4)

plt.legend(fontsize = 12, loc = 'upper center')
plt.tight_layout()
plt.savefig('counts.png')
```

```python
colors = ['#E55807', "#068DA9"]
customPalette = sns.set_palette(sns.color_palette(colors))

df_plot = df_dvlog[['label', 'duration', 'voice_presence', 'face_presence', 'body_presence', 'hand_presence']].reset_index(drop = True)

df_plot['Voice'] = df_plot.apply(lambda x: x['voice_presence'] / x['duration'], axis = 1)
df_plot['Face'] = df_plot.apply(lambda x: x['face_presence'] / x['duration'], axis = 1)
df_plot['Body'] = df_plot.apply(lambda x: x['body_presence'] / x['duration'], axis = 1)
df_plot['Hand'] = df_plot.apply(lambda x: x['hand_presence'] / x['duration'], axis = 1)

melted_df = pd.melt(df_plot, id_vars = ['label'], value_vars = ['Voice', 'Face', 'Body', 'Hand'])

sns.boxplot(x = 'variable', y = 'value', hue = 'label', data = melted_df)

ax = plt.gca()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

ax.tick_params(direction='out', length=10, width=1, color = '#474747', labelcolor = '#474747')
ax.tick_params(which = 'minor', direction='out', length=5, width=1, color = '#474747', labelcolor = '#474747')
ax.set_xlabel("")

ax.set_ylabel("% presence", fontsize=12)

plt.title('D-Vlog Modality Presence')
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] + ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(16)

plt.gcf().set_size_inches(6, 4)

plt.legend(fontsize = 12, loc = 'lower center')

plt.tight_layout()
plt.savefig('presence_fraction.png')
```
